## Confusing scope of variables

In most languages, since

```python
addTen()
```

to work, too, but

```python
# uncomment the code below to see the mistake
#addTen()
```

What's going on?

The assignment to `x` in the second function forces `x` to be
local to the function, therefore its use on the right-hand side is
referencing a local variable that does not have a value yet.  The
first version of the function does not assign to `x` so uses the
reference to `x` defined outside the function.  If you want to
change `x` in the outer environment, then you must set its value
using a returned result, or add a `global` statement.

## Should we remove an element from the beginning or the end of a list?

Which end of a list is it faster to `pop` from?  Remember your linked lists?

Use the ipython
magic function `timeit`.

```python
def testPopFront():
    x = list(range(10000))
    for i in range(100):
        x.pop(0)

def testPopBack():
    x = list(range(10000))
    for i in range(100):
        x.pop()
```

```python
timeit testPopFront()
```

```python
timeit testPopBack()
```

Popping from the back is faster, about twice as fast.

Now, putting the pieces together, the complete python code to find the value of $x$ from a set of 100 random values for which $f(x)$ is closest to zero, we do the following.

```python
import random

def f(x):
    return x**3 - 4*x**2 - 11*x + 30

bestSoFar = (0, f(0))

for step in range(100):
    x = random.uniform(-10, 10)
    fx = f(x)
    if (abs(fx) < abs(bestSoFar[1])):
        bestSoFar = (x, fx)
```

```python
# id imagine it would look something like this. f is our function, remember?
def findRoot(f, xmin, xmax, n=1000):
    bestSoFar = (0, f(0))
    for step in range(n):
        x = random.uniform(xmin, xmax)
        fx = f(x)
        if (abs(fx) < abs(bestSoFar[1])):
            bestSoFar = (x, fx)
    return bestSoFar
```

Now, to collect the `bestSoFar` values.  Let's just build up a list
called `trace` by adding the new value of `bestSoFar` whenever it
is changed.  Can use the `append` list method for this.

    trace.append(bestSoFar)

Here is a function for this. Notice that the function returns two things, the
`bestSoFar` pair, and the `trace`, as a tuple.

```python
import random
import matplotlib.pyplot as plt

def findRoot(f, xmin, xmax, maxSteps=1000):
    bestSoFar = (0, f(0))
    trace = [bestSoFar]
    for step in range(maxSteps):
        # current guess
        x = random.uniform(xmin, xmax)
        # result of passing the current guess to the polynomial function
        fx = f(x)
        # check if the current value of fx is less than the current best fx result so far
        if (abs(fx) < abs(bestSoFar[1])):
          # tuple that holds the current guess of x and the result. if fx is smaller than the best so far, we replace it
            bestSoFar = (x,fx)
          # add each new 'best guess' to trace
            trace.append(bestSoFar)
    return (bestSoFar, trace)
```

The form that is printed when you evaluate it is kind of ugly.  In
python, there are two kinds of `toString` type of methods, for two
purposes:
  * `__repr__` is meant to display a valid python expression that could be used to generate the value
  * `__str__` is meant to display a more human-oriented string that is not meant to be valid python code.

```python
class Node:

    def __init__(self, state, f, g, h):
        self.state = state
        self.f = f
        self.g = g
        self.h = h

    def __repr__(self):
        return 'Node({}, {}, {}, {})'.format(self.state, self.f, self.g, self.h)
```
